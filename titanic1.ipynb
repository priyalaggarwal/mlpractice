```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
feature_cols = ['Pclass', 'Sex', 'Parch']
```

```python
X = train.loc[:, feature_cols]
X.head()
```

```python
y = train.Survived
y.head()
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn import preprocessing
le_sex = preprocessing.LabelEncoder()

#to convert into numbers
train.Sex = le_sex.fit_transform(train.Sex)

#to convert back
#train.Sex = le_sex.inverse_transform(train.Sex)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X, y)
```

```python
test = pd.read_csv('http://bit.ly/kaggletest')
test.head()
```

```python
test.Sex = le_sex.fit_transform(test.Sex)
X_new = test.loc[:, feature_cols]
y_pred = logreg.predict(X_new)
y_pred
```

```python
# to create a dataframe for submission
# we pass a dictionary to the constructor
# But a dictionary is unordered, hence to make sure that PassengerId 
#    is the first column we call another function
x = pd.DataFrame({'PassengerId':test.PassengerId, 'Survived':y_pred}).set_index('PassengerId')
x
```

```python
x.to_csv('titanic_sub.csv')
```
